# Word Embedding using keras

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
##sentence
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
     'I am a good boy',
     'I am a good developer',
     ]
```

```python
sent
```

```python
##vocabilary size
voc_size=10000
```

# One hot representation

```python
one_hot_representation = [one_hot(word,voc_size) for word in sent]
print(one_hot_representation)
```

# Word Embedding Representation

```python
import keras
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
##padding sequencecs
sent_len=10
embedded_docs = pad_sequences(one_hot_representation,maxlen=sent_len,padding='pre')
print(embedded_docs)
```

```python
##dimensions
dim=10
```

```python
##create Embedding layers
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_len))
model.compile(optimizer='adam',loss=keras.losses.mean_squared_error)
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
